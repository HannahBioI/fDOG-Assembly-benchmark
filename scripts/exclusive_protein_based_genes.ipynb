```python
import pandas as pd
```

```python
# dictionary:  busco:{species: int, uniprot: str}
#read in seed genes mapping and ref species
# dictionary: uniprot: busco
seed_dict = {}
seed_uniprot_ids = set()
uniprot_vs_busco = {}
busco_vs_uniprot = {}
with open("../uniprotid_to_group_assignment/seed_species_list_busco.tsv", 'r') as file:
    for line in file:
        line = line.rstrip()
        busco, species = line.split('\t')
        seed_dict[busco] = {"species": species, "uniprot": "", "length": 0, "sum_lcr": 0, "exclusive": 0}
with open("../uniprotid_to_group_assignment/mapping_busco_id_uniport_id.tsv", 'r') as file:
    for line in file:
        line = line.rstrip()
        busco, uniprot = line.split('\t')
        seed_dict[busco]["uniprot"] = uniprot
        uniprot_vs_busco[uniprot] = busco
        busco_vs_uniprot[busco] = uniprot
        seed_uniprot_ids.add(uniprot)
```

```python
#read in domain file
with open("../../results/augustus_busco_core/fdog_assembly_metazoa_core_augustus_new_gallus_forward.domains", "r") as domain_file:
    for line in domain_file:
        seed = False
        line = line.rstrip()
        pairid, orthoid, seqlen, feature, fstart, fend, fweight, fpath, interpro, evalue, bitscore, pstart, pend, plen = line.split("\t")
        busco = pairid.split('#')[0]
        if pairid.split('#')[1] == orthoid:
            seed_dict[busco]['length'] = int(seqlen)
            if feature == 'seg_low_complexity_regions':
                seed_dict[busco]['sum_lcr'] += max(int(fstart), int(fend)) - min(int(fstart), int(fend))
```

```python
with open("exclusive_gene_lists/exclusive_protein_based_pairs_all.tsv", "r") as pairs_file:
    for pair in pairs_file:
        try:
            g1, g2 = pair.rstrip(). split("\t")
        except ValueError:
            print(pair)
            break
        if g1 in seed_uniprot_ids:
            seed = g1
            ortho = g2
        else:
            seed = g2
            ortho = g1
        seed_dict[uniprot_vs_busco[seed]]["exclusive"] += 1
            
```

```python
# make dataframe
df = pd.DataFrame.from_dict(seed_dict, orient='index')
df = df.rename_axis('busco').reset_index()
display(df)
    
```
